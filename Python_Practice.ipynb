I split the dataframe by the average number of losses. I know that the dataframe names are long, but I'm making them explicitly descriptive to cut dow on commenting within the cells.



```python
df['loss'].mean()
```

```python
# Rounding up to 14 losses.
above_mean_loss = (df[df['loss'] > 14]) 
mean_loss_or_less = (df[df['loss'] <= 13])
```

```python
above_mean_loss
```

```python
mean_loss_or_less
```

```python
above_mean_loss['win'].mean()
```

```python
mean_loss_or_less['win'].mean()
```

```python
sci.ttest_ind(above_mean_loss['win'], mean_loss_or_less['win'])
```

```python
plt.hist(df['win'])
plt.axis([0, 30, 0, 12])
plt.show()
```
